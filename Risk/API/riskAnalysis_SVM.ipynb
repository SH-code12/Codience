```python
import pandas as pd
import numpy as np
from sklearn.linear_model import LogisticRegression
```

```python
train = pd.read_csv('/Users/rebuy/Desktop/FCAI/GP/Research Papers/Risk Analysis/5907002/apachejit/dataset/apachejit_train.csv')
train.head()
```

# Pre Processing

```python
X_train= train.drop(['buggy','commit_id','project','fix','year','author_date',"fix"],axis=1)
y_train= train['buggy']
X_train
```

```python
y_train
```

# Model Training

## SVM

```python
from sklearn.svm import SVC
from sklearn.model_selection import GridSearchCV

# Define parameter grid
param_grid = {
    'C': [0.1, 1, 10, 50],            # Regularization strength
    'kernel': ['linear', 'rbf', 'poly'],  # Kernel type
    'gamma': ['scale', 'auto'],        # Kernel coefficient for 'rbf' and 'poly'
    'degree': [2, 3, 4]                # Only used for 'poly'
}

# Create SVM
svc = SVC(probability=True)  # probability=True to get predicted probabilities

# Grid search with 5-fold CV
grid_search = GridSearchCV(svc, param_grid, cv=5, scoring='f1', n_jobs=-1, verbose=2)
grid_search.fit(X_train, y_train)
```

```python
print("Best parameters found:", grid_search.best_params_)
print("Best F1 score (CV):", grid_search.best_score_)
```
