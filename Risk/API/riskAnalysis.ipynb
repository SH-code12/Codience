```python
import pandas as pd
import numpy as np
from sklearn.linear_model import LogisticRegression
```

```python
train = pd.read_csv('./Dataset/apacheJIT/apachejit_train.csv')

train.head()
```

# Pre Processing

```python
X_train= train.drop(['commit_id','project','fix','year','author_date',"fix"],axis=1)
y_train= train['buggy']
X_train
```

```python
y_train
```

# Model Training

```python
from sklearn.linear_model import LogisticRegression

# create the model
model = LogisticRegression()

# train the model
model.fit(X_train, y_train)
```

# Testing

```python
test = pd.read_csv('./Dataset/apacheJIT/aapachejit_test_large.csv')
test
```

```python
X_test= test.drop(['commit_id','project','fix','year','author_date',"fix"],axis=1)
y_test= test['buggy']
```

```python
model.score(X_test, y_test)
```

# Saving the Model

```python
import pickle

# save the trained model
with open("risk_logistic_model.pkl", "wb") as f:
    pickle.dump(model, f)
```
